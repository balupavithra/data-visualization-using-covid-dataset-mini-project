```python
 #bar plot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('country_wise_latest.csv')

# Inspect data
print("Columns in the DataFrame:", data.columns)
print("First few rows of the DataFrame:\n", data.head())

# Sort the data to plot the top 10 countries by confirmed cases
top_countries = data.nlargest(10, 'Confirmed')

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Confirmed', y='Country/Region', data=top_countries)
plt.title('Top 10 Countries by Confirmed COVID-19 Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Country/Region')
plt.show()
```

```python
#scatter plot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('country_wise_latest.csv')

# Inspect data
print("Columns in the DataFrame:", data.columns)
print("First few rows of the DataFrame:\n", data.head())

# Create a scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Confirmed', y='Deaths', hue='WHO Region', data=data)
plt.title('Confirmed Cases vs Deaths by WHO Region')
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.legend(title='WHO Region')
plt.show()
```

```python
#heatmap correlation
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
file_path = 'country_wise_latest.csv'
data = pd.read_csv(file_path)

# Inspect data
print("Columns in the DataFrame:", data.columns)
print("First few rows of the DataFrame:\n", data.head())

# Select only numeric columns
numeric_data = data.select_dtypes(include=[float, int])

# Compute correlation matrix
corr_matrix = numeric_data.corr()

# Create a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of COVID-19 Statistics')
plt.show()
```

```python
#histogram
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('country_wise_latest.csv')

# Histogram of Confirmed Cases
plt.figure(figsize=(12, 6))
sns.histplot(data['Confirmed'], bins=30, kde=True)
plt.title('Distribution of Confirmed COVID-19 Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Frequency')
plt.show()
```

```python
#Box Plot for Deaths by WHO Region
# Box Plot for Deaths by WHO Region
plt.figure(figsize=(12, 6))
sns.boxplot(x='WHO Region', y='Deaths', data=data)
plt.title('Deaths by WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Deaths')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pair Plot for Selected Columns
# Select columns for the pair plot
selected_columns = ['Confirmed', 'Deaths', 'Recovered', 'Active']
pair_data = data[selected_columns]

# Pair Plot
sns.pairplot(pair_data)
plt.suptitle('Pair Plot for Confirmed, Deaths, Recovered, and Active Cases', y=1.02)
plt.show()
```

```python
#Line Plot for Weekly Increase
# Top 5 countries by confirmed cases
top_5_countries = data.nlargest(5, 'Confirmed')

plt.figure(figsize=(12, 6))
for country in top_5_countries['Country/Region']:
    country_data = data[data['Country/Region'] == country]
    plt.plot(country_data['1 week % increase'], label=country)

plt.title('Weekly Increase in Confirmed Cases')
plt.xlabel('Country')
plt.ylabel('1 Week % Increase')
plt.legend(title='Country')
plt.show()
```

```python

# Violin Plot for Recovered Cases by WHO Region
plt.figure(figsize=(12, 6))
sns.violinplot(x='WHO Region', y='Recovered', data=data)
plt.title('Recovered Cases by WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Recovered Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Swarm Plot for Active Cases by WHO Region
plt.figure(figsize=(12, 6))
sns.swarmplot(x='WHO Region', y='Active', data=data)
plt.title('Active Cases by WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Active Cases')
plt.xticks(rotation=45)
plt.show()
```
